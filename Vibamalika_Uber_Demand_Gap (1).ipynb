# **Project Summary -**

### Project Summary: Uber Supply-Demand Gap Analysis

This project conducts an Exploratory Data Analysis (EDA) on a dataset from a ride-hailing service, simulating a scenario similar to Uber, to identify the root causes behind the "Supply-Demand Gap." The primary business objective is to understand why a significant number of ride requests are either cancelled or result in "No Cars Available," leading to poor customer experience and lost revenue.

The analysis begins with data loading and a thorough inspection of the dataset, which contains over 6,700 ride requests with details like pickup point (Airport or City), status, driver ID, and timestamps. Initial data cleaning involves handling missing values in 'Driver id' and 'Drop timestamp' columns, preparing the data for robust analysis.

The core of the analysis is a structured visualization process following the UBM (Univariate, Bivariate, Multivariate) rule. Univariate analysis reveals the overall distribution of requests, cancellations, and completed trips, showing a high proportion of "Trip Completed" status but also a significant number of "No Cars Available" events. Bivariate analysis explores relationships between variables, such as the distribution of request status by pickup location. This analysis uncovers a critical insight: a higher number of rides are requested from the City, but a larger proportion of cancellations and "No Cars Available" incidents occur at the Airport, highlighting a supply-demand mismatch at this key location.

Further investigation into the time dimension reveals the pattern of this mismatch. Time-series analysis of requests and drops shows peak demand periods. By visualizing the distribution of request statuses over time, the analysis identifies specific time slots and days when the supply-demand gap is most pronounced, primarily at the Airport during peak hours. This suggests that driver allocation is insufficient to meet the surge in demand at these times.

The analysis also explores other factors. A correlation heatmap indicates a weak or non-existent linear relationship between 'Request id' and 'Driver id', suggesting that driver assignment may not be optimally matched to request patterns. The code also delves into individual driver performance by plotting top drivers by request ID, which could be a preliminary step in identifying high-performing drivers, but may also point to potential over-utilization of a small pool of drivers, exacerbating shortages when demand spikes.

The key findings point to a systemic problem where the supply of drivers is not aligned with the spatial and temporal distribution of demand. The high cancellation rate at the Airport during peak hours suggests that either drivers are unwilling to accept these trips (perhaps due to distance or traffic) or there are simply not enough of them available to meet the demand. The analysis suggests that the company should focus on implementing dynamic driver incentives, such as surge pricing or bonuses, specifically targeted at Airport pickups during identified peak times to increase supply. Improving driver allocation algorithms to better predict and respond to real-time demand at strategic locations like the Airport is also a recommended solution. Finally, understanding the reasons behind cancellations (e.g., driver or passenger-initiated) could help in refining operational strategies to reduce this metric and improve overall efficiency. The project concludes that a data-driven approach is essential for optimizing driver distribution and balancing supply and demand to enhance customer satisfaction and revenue.

### Variables Description

- **Request id**
  - **Type**: Integer
  - **Description**: A unique identifier for each ride request made by a passenger. This serves as the primary key for the dataset and helps track individual trip records.

- **Pickup point**
  - **Type**: Categorical (Object)
  - **Description**: Indicates the location from which the ride was requested. The dataset contains two categories:
    - **Airport**: Requests originating from the airport area.
    - **City**: Requests originating from within the city.
  - This variable is crucial for analyzing location-based supply-demand gaps.

- **Driver id**
  - **Type**: Float
  - **Description**: A unique identifier for the driver assigned to the request. This field contains **2,650 missing values**, indicating instances where no driver was available or assigned. This is a key indicator of supply-side shortages.

- **Status**
  - **Type**: Categorical (Object)
  - **Description**: Represents the outcome of each ride request. It has three possible values:
    - **Trip Completed**: The ride was successfully fulfilled.
    - **Cancelled**: The ride was cancelled, either by the driver or the passenger.
    - **No Cars Available**: No driver was available to accept the request.
  - This is the **target variable** for identifying the supply-demand gap.

- **Request timestamp**
  - **Type**: Object (Datetime)
  - **Description**: The date and time when the ride request was made. This variable enables time-series analysis to identify peak demand hours, days, and temporal patterns of service failures.

- **Drop timestamp**
  - **Type**: Object (Datetime)
  - **Description**: The date and time when the ride was completed. This field contains **3,914 missing values**, corresponding to rides that were not completed (cancelled or no cars available). This variable helps in analyzing trip duration and completion rates.

#### Chart - 1

````python
# Chart - 1 visualization code
plt.style.use('dark_background')
md.columns=md.columns.str.strip()
plt.figure(figsize=(12,8))
ax=sns.countplot(x='Status',data=md,hue="Status",palette="dark",legend=0)
plt.title('Count for each status type',fontsize=12)
plt.xlabel('Status',fontsize=12)
plt.ylabel('Count',fontsize=12)
for p in ax.patches:
    count=int(p.get_height())
    x=p.get_x()+p.get_width()/2
    y=p.get_height()
    ax.annotate(
        str(count),
        (x,y),
        ha='center',
        va='bottom',
        fontsize=12,
        color='white',
        fontweight='bold'
    )
plt.grid(False)
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

I chose a count plot (bar chart) because it is the most effective way to compare the frequency of categorical variables—in this case, the three Status types (Trip Completed, Cancelled, No Cars Available). The bar chart allows for a clear, visual comparison of volumes across categories, making it easy to identify which status dominates and which is the biggest problem area.

##### 2. What is/are the insight(s) found from the chart?

The chart reveals that Trip Completed has the highest count, but a combined total of Cancelled and No Cars Available indicates a significant supply-demand gap. The No Cars Available status is particularly high, suggesting that driver supply is insufficient to meet demand, leading to lost revenue and poor customer experience.
